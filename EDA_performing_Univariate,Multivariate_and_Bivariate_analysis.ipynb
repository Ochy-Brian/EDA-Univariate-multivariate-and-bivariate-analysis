#Univariate analysis-analysing data that consist of only one variable

```python
#find aggregate sales among branches
sns.countplot(data['Branch'])
```

```python
data['Branch'].value_counts().plot(kind="pie",autopct="%1.2f%%")
```

```python
#find the most popular payment method used by customers(categorical column)
data['Payment'].value_counts().plot(kind='bar')
```

```python
#3.Find the distribution of customer ratings
sns.distplot(data['Rating'])
```

```python
#find skewness and if its zero the distribution is almost normal
data['Rating'].skew()
```

```python
#find the distribution of the cost of goods sold
sns.distplot(data['cogs'])
```

```python
#skewness means of data is postive its right skewed if data is negative its left skewd
data['cogs'].skew()
```

#Bivariate analysis/multivariate analysis

```python
#does the cost of goods sold affect the ratings that the customers provide?
sns.scatterplot(x=data['cogs'],y=data['Rating'])
```

```python
#does gross income affect the ratings that the customers provide
sns.scatterplot(x=data['gross income'], y=data['Rating'])
```

```python
#find the most profitable branch as per gross income
sns.barplot(x=data['Branch'],y=data['gross income'])
```

```python
sns.barplot(x=data['Branch'],y=data['gross income'],hue=data['City'])
```

```python
#is there any relationship between gender and gross income(bivairiate analysis)
sns.boxplot(x=data['Gender'],y=data['gross income'])
```

```python
#multivariate analysis
sns.boxplot(x=data['Gender'],y=data['gross income'],hue=data['Customer type'])
```

```python
#find the product line that generates the most income
data.columns
```

```python
sns.barplot(x=data['Product line'],y=data['gross income'])
plt.xticks(rotation=80)
plt.show()
```

```python
#find the highest unit price in the product line
sns.barplot(x=data['Product line'],y=data['Unit price'])
```

```python
sns.barplot(x=data['Product line'],y=data['Unit price'])
plt.xticks(rotation=80)
plt.show()
```

```python
#find different payment methods used by customers citywise
pd.crosstab(data['City'],data['Payment'])
```

```python
#checking relationships between two categorical columns
sns.heatmap(pd.crosstab(data['City'],data['Payment']))
```

```python
#which productline is purchased in the highest quantity
(data.groupby('Product line').sum()['Quantity']).plot(kind='bar')
```
